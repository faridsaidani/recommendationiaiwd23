```python
import pandas as pd
import numpy as np

ratings_df = pd.read_json("ecommerce/ratings.json")
customers_df = pd.read_json("ecommerce/customers.json")
products_df = pd.read_json("ecommerce/products.json")
```

```python
products_df = products_df.drop(['UnitPrice', 'Id'], axis=1)
```

```python
products_df['Name'] = products_df['Name'].str.lower()
```

```python
products_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(products_df['Name'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(vectorized)
```

```python
print(similarities)
```

```python
df = pd.DataFrame(similarities, columns=products_df['Name'], index=products_df['Name']).reset_index()

df.head()
```

```python
def product_recommendation(product):
  recommendations = pd.DataFrame(df.nlargest(11,product)['Name'])
  recommendations = recommendations[recommendations['Name']!=product]
  return (recommendations)
```

```python
print(product_recommendation('xiaomi redmi note 6 pro 32gb'))
```
